```python
import pandas as pd
```

```python
dataset = pd.read_csv("Dataset\Steel_industry_data.csv")
dataset['date']
```

```python
dataset['date'] = pd.to_datetime(dataset['date'], format='%d/%m/%Y %H:%M')
dataset.set_index('date', inplace=False)
dataset
```

```python
import matplotlib.pyplot as plt

# Create a time series plot
plt.figure(figsize=(12, 6))
plt.plot(dataset.index, dataset['Usage_kWh'])

# Configure the x-axis to display the date labels
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(prune='both'))


plt.title('Electricity Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Usage_kWh')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.show()
```
